```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

def display_image(array):
    plt.imshow(array, cmap='gray')
    plt.axis('off')
    plt.show()

def read_image(path):
    original = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    print("Original Image: ", path)
    display_image(original)
    return original

def save_image(array, path, file_name):
    if array.dtype != np.uint8:
        array = ((array - array.min()) / (array.max() - array.min()) * 255).astype('uint8')

    image = Image.fromarray(array)

    if not os.path.exists(path):
        os.makedirs(path)

    full_path = os.path.join(path, file_name)
    image.save(full_path, 'PNG')
    print(f"Image saved at {full_path}")

def bilateral_filtering(image_matrix,region):
    filtered = cv2.bilateralFilter(image_matrix, d=-1, sigmaColor=region, sigmaSpace=region)
    print("Filtered Image: ")
    display_image(filtered)
    return filtered
```

# Image 3.png / Region Size 35

```python
image_3 = read_image('Original_Images/Speckle_images/3.png')
image_3_filtered = bilateral_filtering(image_3, region=35)
save_image(image_3_filtered, 'Results/Speckle_Filtering_Results', 'image_3_filtered.png')
```
